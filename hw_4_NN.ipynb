# Задание 1

обучить рукописную сверточную нейронную сеть (с падением размера ядра свертки и последовательностью блоков свертка-пулинг (conv-pool)-(conv-pool)-...) на датасете cifar-10



```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
import os
```

```python
classes = ['самолет', 'автомобиль', 'птица', 'кот', 'олень', 'собака', 'лягушка', 'лошадь', 'корабль', 'грузовик']
```

```python
# установка параметров нейросети
batch_size = 512
num_classes = 10
epochs = 5
data_augmentation = False
num_predictions = 20
save_dir = os.path.join(os.getcwd(), 'saved_models')
model_name = 'keras_cifar10_trained_model.h5'

# разделение тренировочной и тестовой выборки
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'тренировочные примеры')
print(X_test.shape[0], 'тестовые примеры')

# преобразование матрицы чисел 0-9 в бинарную матрицу чисел 0-1
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)


X_train = X_train / 255.0
X_test = X_test / 255.0
```

строим архитектуру модели

```python
model_1 = Sequential([
    Conv2D(16, (5, 5), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(32, (3, 3), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(256, activation='relu', input_shape=(13,)),
    Dense(64, activation='linear'),
    Dense(num_classes, activation='softmax'),
])

model_1.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_1.summary()
```

```python
model_1.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(X_test, y_test),
          shuffle=True)
```

# Задание 2
оценить рост точности при увеличении ширины сети (больше фильтров)

Увеличим количество каналов на первом конволюционном слое с 32 до 64, на втором conv слое с 16 до 128

```python
model_2 = Sequential([
    Conv2D(64, (5, 5), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(256, activation='relu', input_shape=(13,)),
    Dense(64, activation='linear'),
    Dense(num_classes, activation='softmax'),
])

model_2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_2.summary()
```

```python
model_2.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(X_test, y_test),
          shuffle=True)
```

#Вывод:

Благодаря увеличению количества каналов, метрика выросла: на тесте с 0.6169 до 0.6828

# Задание 3

оценить рост точности при увеличении глубины сети (больше слоев)

добавим дополнительные слои конволюции и пулинга в сеть

```python
model_3 = Sequential([
    Conv2D(16, (5, 5), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(32, (3, 3), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    #ниже дополнительные слои конволюции и пулинга
    Conv2D(64, (3, 3), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), padding='same', input_shape=X_train.shape[1:], activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    
    Flatten(),
    Dense(256, activation='relu', input_shape=(13,)),
    Dense(64, activation='linear'),
    Dense(num_classes, activation='softmax'),
])

model_3.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model_3.summary()
```

```python
model_3.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(X_test, y_test),
          shuffle=True)
```

# Вывод

# Дополнительно

для себя попробовал построить модель с использованием готовой архитектуры ResNet50

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Dense, Flatten, GlobalAvgPool2D, Dropout, BatchNormalization
from keras.models import Model
from keras.callbacks import ModelCheckpoint

from keras.applications.resnet import ResNet50, preprocess_input as resnet_preproc
from keras.applications.vgg16 import VGG16, preprocess_input as vgg_preproc

import matplotlib.pyplot as plt
```

```python
# разделение тренировочной и тестовой выборки
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'тренировочные примеры')
print(X_test.shape[0], 'тестовые примеры')

# преобразование матрицы чисел 0-9 в бинарную матрицу чисел 0-1
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
# необходимый препроцессинг данных для ResNet
X_train = resnet_preproc(X_train)
X_test = resnet_preproc(X_test)
```

```python
input_shape = (32, 32)
```

В качестве основы сети используем предобученную сеть ResNet50, убираем только последний полносвязный слой

```python
base_model = ResNet50(include_top=False, weights='imagenet', input_shape=(input_shape[0], input_shape[1], 3))
base_model.trainable = False
base_model.summary()
```

добавляем к resnet следующие слои и дообучаемся только на этих слоях:

```python
x = GlobalAvgPool2D()(base_model.output)
x = Dropout(0.1)(x)
x = BatchNormalization()(x)
x = Dense(512, activation='relu')(x)
x = Dense(num_classes, activation='softmax')(x)

model_4 = Model(inputs=base_model.input, outputs=x)
model_4.summary()
```

```python
model_4.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model_4.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(X_test, y_test),
          shuffle=True)
```
